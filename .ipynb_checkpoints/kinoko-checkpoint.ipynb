```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
```

```python
data = pd.read_csv("kinoko.csv")
```

```python
chuuku = data[data["柄の断面積(cm2)"]=="A"].copy()
tsumari = data[data["柄の断面積(cm2)"]=="B"].copy()
fuwafuwa = data[data["柄の断面積(cm2)"]=="C"].copy()
```

```python
chuuku["断面積(cm2)"]=np.pi / 4 * (chuuku["柄の外径(mm)"]**2 - chuuku["柄の内径"]**2)
tsumari["断面積(cm2)"]= np.pi / 4 * (tsumari["柄の外径(mm)"]**2)
fuwafuwa["断面積(cm2)"]= np.pi / 4 * (fuwafuwa["柄の外径(mm)"]**2 - fuwafuwa["柄の内径"]**2)
```

```python
x_param="柄の外径(mm)"
y_param="傘の重さ(g)"

x=np.log(chuuku[x_param])
y=np.log(chuuku[y_param])
plt.scatter(x,y,color="blue")
X=sm.add_constant(x)
model=sm.OLS(y,X)
fitted=model.fit()
x_pred = np.linspace(x.min(),x.max(),50)
X_pred = sm.add_constant(x_pred)
y_pred = fitted.predict(X_pred)
plt.plot(x_pred,y_pred,"blue")
print("Slope for chuuku:",fitted.params[1])

x=np.log(tsumari[x_param])
y=np.log(tsumari[y_param])
plt.scatter(x,y,color="green")
X=sm.add_constant(x)
model=sm.OLS(y,X)
fitted=model.fit()
x_pred = np.linspace(x.min(),x.max(),50)
X_pred = sm.add_constant(x_pred)
y_pred = fitted.predict(X_pred)
plt.plot(x_pred,y_pred,"green")
print("Slope for tsumari:",fitted.params[1])

x=np.log(fuwafuwa[x_param])
y=np.log(fuwafuwa[y_param])
plt.scatter(x,y,color="orange")
X=sm.add_constant(x)
model=sm.OLS(y,X)
fitted=model.fit()
x_pred = np.linspace(x.min(),x.max(),50)
X_pred = sm.add_constant(x_pred)
y_pred = fitted.predict(X_pred)
plt.plot(x_pred,y_pred,"orange")
print("Slope for fuwafuwa:",fitted.params[1])
```

```python
data_for_moment = data.copy()
data_for_moment["柄の内径"]=data_for_moment["柄の内径"].fillna(0)
data_for_moment["断面２次モーメント"] = np.pi / 2 * ((data_for_moment["柄の外径(mm)"]/2)**4 - (data_for_moment["柄の内径"]/2)**4)
```

```python
x_param = "断面２次モーメント"
y_param = "傘の重さ(g)"

x=np.log(data_for_moment[data_for_moment["柄の断面積(cm2)"]=="A"].dropna(subset=[x_param,y_param])[x_param][data_for_moment[data_for_moment["柄の断面積(cm2)"]=="A"].dropna(subset=[x_param,y_param])[x_param]>0])
y=np.log(data_for_moment[data_for_moment["柄の断面積(cm2)"]=="A"].dropna(subset=[x_param,y_param])[y_param][data_for_moment[data_for_moment["柄の断面積(cm2)"]=="A"].dropna(subset=[x_param,y_param])[x_param]>0])
plt.scatter(x,y,color="blue")
X=sm.add_constant(x)
model=sm.OLS(y,X)
fitted=model.fit()
x_pred = np.linspace(x.min(),x.max(),50)
X_pred = sm.add_constant(x_pred)
y_pred = fitted.predict(X_pred)
plt.plot(x_pred,y_pred,"blue")
print("Slope for chuuku:",fitted.params[1])

x=np.log(data_for_moment[data_for_moment["柄の断面積(cm2)"]=="C"].dropna(subset=[x_param,y_param])[x_param][data_for_moment[data_for_moment["柄の断面積(cm2)"]=="C"].dropna(subset=[x_param,y_param])[x_param]>0])
y=np.log(data_for_moment[data_for_moment["柄の断面積(cm2)"]=="C"].dropna(subset=[x_param,y_param])[y_param][data_for_moment[data_for_moment["柄の断面積(cm2)"]=="C"].dropna(subset=[x_param,y_param])[x_param]>0])
plt.scatter(x,y,color="green")
X=sm.add_constant(x)
model=sm.OLS(y,X)
fitted=model.fit()
x_pred = np.linspace(x.min(),x.max(),50)
X_pred = sm.add_constant(x_pred)
y_pred = fitted.predict(X_pred)
plt.plot(x_pred,y_pred,"green")
print("Slope for tsumari:",fitted.params[1])

x=np.log(data_for_moment[data_for_moment["柄の断面積(cm2)"]=="B"].dropna(subset=[x_param,y_param])[x_param][data_for_moment[data_for_moment["柄の断面積(cm2)"]=="B"].dropna(subset=[x_param,y_param])[x_param]>0])
y=np.log(data_for_moment[data_for_moment["柄の断面積(cm2)"]=="B"].dropna(subset=[x_param,y_param])[y_param][data_for_moment[data_for_moment["柄の断面積(cm2)"]=="B"].dropna(subset=[x_param,y_param])[x_param]>0])
plt.scatter(x,y,color="orange")
X=sm.add_constant(x)
model=sm.OLS(y,X)
fitted=model.fit()
x_pred = np.linspace(x.min(),x.max(),50)
X_pred = sm.add_constant(x_pred)
y_pred = fitted.predict(X_pred)
plt.plot(x_pred,y_pred,"orange")
print("Slope for fuwafuwa:",fitted.params[1])
```

```python
data_for_pca = data[["傘の重さ(g)","柄の長さ(mm)","柄の外径(mm)","柄の断面積(cm2)"]]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(data_for_pca[["傘の重さ(g)","柄の長さ(mm)","柄の外径(mm)"]])
pca_chuuku = pca.transform(data_for_pca[["傘の重さ(g)","柄の長さ(mm)","柄の外径(mm)"]][data_for_pca["柄の断面積(cm2)"]=="A"])[:,(0,1)]
pca_tsumari = pca.transform(data_for_pca[["傘の重さ(g)","柄の長さ(mm)","柄の外径(mm)"]][data_for_pca["柄の断面積(cm2)"]=="B"])[:,(0,1)]
pca_fuwafuwa = pca.transform(data_for_pca[["傘の重さ(g)","柄の長さ(mm)","柄の外径(mm)"]][data_for_pca["柄の断面積(cm2)"]=="C"])[:,(0,1)]

plt.scatter(pca_chuuku[:,0],pca_chuuku[:,1],color="blue")
plt.scatter(pca_tsumari[:,0],pca_tsumari[:,1],color="green")
plt.scatter(pca_fuwafuwa[:,0],pca_fuwafuwa[:,1],color="orange")

```

```python
pca_chuuku_only = PCA()
pca_chuuku_only.fit(data_for_pca[["傘の重さ(g)","柄の長さ(mm)","柄の外径(mm)"]][data_for_pca["柄の断面積(cm2)"]=="A"])
pca_chuuku_only_result = pca_chuuku_only.transform(data_for_pca[["傘の重さ(g)","柄の長さ(mm)","柄の外径(mm)"]][data_for_pca["柄の断面積(cm2)"]=="A"])[:,(0,1)]
plt.scatter(pca_chuuku_only_result[:,0],pca_chuuku_only_result[:,1])
```
